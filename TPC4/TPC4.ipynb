```python
print('''
    -----------------------
    Menu de opções:
        (1) Criar Lista 
        (2) Ler Lista
        (3) Soma
        (4) Média
        (5) Maior
        (6) Menor
        (7) estaOrdenada por ordem crescente
        (8) estaOrdenada por ordem decrescente
        (9) Procura um elemento
        (0) Sair
    -----------------------
     ''')
import random
lista = []
opc= int(input("Escolha a opção de 0 a 9 do menu"))
while opc != 0:
    if opc == 1:
        N = int(input("introduza o numero de elementos da lista"))
        lista = []
        i = 0
        while i < N:
            num = random.randint(1, 100)
            lista.append(num)
            i = i + 1
        print("Lista criada: {}".format(lista))

    if opc == 2:
        N = int(input("introduza o numero de elmentos da lista"))
        lista = []
        i = 0
        while N > i:
            num = int(input("Introduza os valores da lista"))
            i = i + 1
            lista.append(num)
        print("Lista criada: {}".format(lista))  

    if opc == 3:
        soma = 0
        for n in lista:
            soma = soma + n 
        print("A soma da lista {} é {}".format(lista, soma)) 

    if opc == 4:
        soma = 0
        media = 0
        for n in lista:
            soma = soma + n
            n = n + 1
        media = (soma) / len(lista)
        print("A média da lista é {}".format(media)) 
    
    if opc == 5:
        maior = lista[0]
        for n in lista[1:]:
            if n > maior:
                maior = n 
        print("O maior numero da lista é {}".format(maior))
    
    if opc == 6:
        menor = lista[0]
        for n in lista[1:]:
            if n < menor:
                menor = n 
        print("O menor numero da lista é {}".format(menor))
    
    if opc == 7:
        if sorted(lista) == lista:
            res = "Sim"
        else:
            res = "Não"
        print("A lista está ordenada por ordem crescente?" , res)
    
    if opc == 8:
        if sorted(lista, reverse=True) == lista:
            res = "Sim"
        else: 
            res = "Não"
        print("A lista está ordenada por ordem decrescente?" , res)
    
    if opc == 9:
        print(lista)
        n = int(input("Introduza um número para procurar"))
        if n in lista:
            print("A posição do número é, #{}".format(lista.index(n)))
        else:
            print('O número procurado não está na lista')


    opc= int(input("Escolha a opção de 0 a 9 do menu"))

print('A lista final é: ', lista)
    
    
```
