```python
def criaPol():
    polinomio = []
    grau = int(input("Qual é o grau do polinomio"))
    while grau >= 0:
        coeficiente = int(input(f"Qual é o coefciente de grau {grau}"))
        polinomio.append((coeficiente, grau))
        grau = grau - 1
    return polinomio 
#Criar polinomio



def lePolinomios(fnome):
    file = open(fnome, "r")
    for line in  file:
        print(line, end="")
    return 
#Ler Polinomio

def pot(base,exp):
    res = 1
    while exp > 1:
        res = res * base
        exp = exp - 1

def calcPolinomio(p, x):
    res = 0
    for monomio in p:
        coeficiente, grau = monomio
        res = res + pot(x,grau)*coeficiente
    return res
#Calcular Polinomio


def grauPolinomios(listap):
    maior = -1
    for polinomio in listap:
        for monomio in polinomio:
            coeficiente, grau = monomio
            if grau > maior:
                maior = grau 
    return maior
#Maior grau de polinomio


def calcular_derivada(polinomio):
    derivada = []
    for monomio in polinomio:
        coeficiente, grau = monomio
        if coeficiente and grau != 0:
            coeficiente = coeficiente * grau 
            grau = grau - 1
            derivada.append((coeficiente, grau))
    return derivada 
#Derivada


def calcular_soma(p1, p2):
    soma = []
    for monomio1 in p1:
        for monomio2 in p2: 
            if monomio1[1] == monomio2[1]:
                coef = monomio1[0] + monomio2[0]
                soma.append((coef,monomio1[1]))
    return soma 
#Soma


import matplotlib.pyplot as plt
import numpy as np

def plotar_polinomio_grafico(polinomio):
    x = np.linspace(-100, 100, 400)  # Gera pontos no eixo x
    y = np.zeros_like(x)
    for monomio in polinomio:
        coeficiente, grau = monomio
        y = y + (coeficiente * (x ** grau))
    plt.plot(x, y)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Gráfico do Polinômio')
    plt.grid(True)
    plt.show()
#Gráfico

def gravar_polinomios(fnome, lista_de_polinomios):
    with open(fnome, 'w') as arquivo:
        for polinomio in lista_de_polinomios:
            arquivo.write(str(polinomio) + '\n')
#gravar polinomios

def listar_polinomios(lista_de_polinomios):
    for polinomio in lista_de_polinomios:
        print(polinomio)
#listar

```
